# Exploratory Analysis

```python
import pathlib

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd

%matplotlib inline
```

## Basic Statistics of Training Data

```python
training_data = pd.read_csv(pathlib.Path.cwd() / '../data/train_label.csv' )
training_data.sample(n=10)
```

```python
training_data.describe()
```

```python
labels = training_data["Label"].unique()
training_data.groupby("Label").count()
```

```python
fig = plt.figure(figsize=(12,12))

for index, label in enumerate(sorted(labels)):
    ax = fig.add_subplot(len(labels), 1, index + 1)
    sample = training_data.loc[training_data["Label"] == label].sample()
    image = mpimg.imread(pathlib.Path.cwd() / f'../data/train_image/train_image/{label}.png')
    plt.imshow(image)
    ax.set_title(f"{label} - {sample['ID'].iloc[0]}")
```
